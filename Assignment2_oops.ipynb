```python
#Q3
match_dict = {"test1": {'Dhoni': 56, "Balaji": 85}, 'test2': {'Dhoni': 87, 'Balaji': 200}}

def Max_Score(d):
    total_runs = {}
    for k in d.keys():
        for n in d[k].keys():
            if n in total_runs.keys():
                total_runs[n] = total_runs[n] + d[k][n]
            else:
                total_runs[n] = d[k][n]
    print("Total Run Scored by Each Playes in 2 Tests: ")
    print(total_runs)

    print("Player With highest score")
    maxruns = -1
    for n in total_runs.keys():
        if total_runs[n] > maxruns:
            maxname = n
            maxruns = total_runs[n]

    return (maxname, maxruns)

summary = Max_Score(match_dict)
print(summary)
```

```python
#Q12-1
class Isoceles_triangle:
    
    def __init__(self, angle1, angle2, angle3):
        
        self.angle1 = angle1
        self.angle2 = angle2
        self.angle3 = angle3
    
    def iso_triangle(self):
        
        if self.angle1 == self.angle2 and self.angle1 != self.angle3 and self.angle2 != self.angle3:
            
            print("It is an isoceles triangle")
        
        elif self.angle2 == self.angle3 and self.angle2 != self.angle1 and self.angle3 != self.angle1:
            
            print("It is an isoceles triangle")
        
        elif self.angle1 == self.angle3 and self.angle1 != self.angle2 and self.angle3 != self.angle2:
            
            print("It is an isoceles triangle")
        
        else:
            
            print("It is not an isoceles triangle")


class right_triangle:
    
    def __init__(self, angle1, angle2, angle3):
        
        self.angle1 = angle1
        self.angle2 = angle2
        self.angle3 = angle3
    
    def r_triangle(self):
        
        if self.angle1 == 90 or self.angle2 == 90 or self.angle3 == 90:
            
            print("It is a right angled triangle")
        
        else:
            
            print("It is not a right angled triangle")

            
            
class isoceles_right_triangle(Isoceles_triangle, right_triangle):
    
    def iso_right_triangle(self):
        
        if self.angle1 == 90 and self.angle2 == self.angle3:
            
            print("It is an isoceles right angled triangle")
        
        elif self.angle2 == 90 and self.angle1 == self.angle3:
            
            print("It is an isoceles right angled triangle")
        
        elif self.angle3 == 90 and self.angle1 == self.angle2:
            
            print("It is an isoceles right angled triangle")
        
        else:
            
            print("It is not an isoceles right angled triangle")
            
obj_6 = isoceles_right_triangle(90,45,45)
obj_6.iso_triangle()
obj_6.r_triangle()
obj_6.iso_right_triangle()
```
